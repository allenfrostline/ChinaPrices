```python
import cmocean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
import numpy as np
import seaborn as sns
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['axes.facecolor']='#fafafa'
zh_font = font_manager.FontProperties(fname=r'/System/Library/Fonts/PingFang.ttc', size=10)
pd.set_option('display.max_rows', 10)
```

```python
filelist = ['KTV',
            'SPA',
            '健身',
            '按摩',
            '洗浴',
            '洗衣',
            '理发',
            '美容',
            '足浴',
            '汽车服务',
            '美甲美睫',]
fn = len(filelist)
```

```python
bidlist = ['闸北',
           '虹口',
           '普陀',
           '杨浦',
           '嘉定',
           '宝山',
           '黄埔',
           '奉贤',
           '长宁',
           '闵行',
           '松江',
           '卢湾',
           '浦东',
           '静安',
           '徐汇',
           '崇明',
           '青浦',
           '金山']
bn = len(bidlist)
```

```python
df = [pd.read_csv(f+'.csv') for f in filelist]
```

<div style="font-family:PingFang SC; font-size:12pt">
经过初步的筛选，我们总结了十项具体服务，同时满足以下三条标准：
<ul>
<li> 文娱生活：该服务属于大多数中产阶级日常涉及的文娱与个人服务；</li>
<li> 较为普遍：该服务在美团上的总销量为对应子类别前五；</li>
<li> 跨省可比：该服务不是上海特有的。</li>
</ul>

<h3 style="font-family:PingFang SC; font-size:12pt"> 1. KTV（3小时）</h3>

```python
%config InlineBackend.figure_format = 'retina'

tmp1 = df[0].服务.str.find('三小时').apply(lambda x: max(x,0))
tmp2 = df[0].服务.str.find('3小时').apply(lambda x: max(x,0))
tmp = tmp1 + tmp2
KTV = df[0].ix[tmp[tmp!=0].index]
quan = sum(KTV.销量)

营收 = KTV.价格 * KTV.销量
营收.name = '营收'
KTV = pd.concat([KTV, 营收], axis=1)
bid = KTV.groupby('区域', as_index=True).sum()
bidincm = bid.营收
bidquan = bid.销量
bidprce = bidincm / bidquan
bid = pd.concat([bidprce, bidquan], axis=1).reindex(bidlist)
bid.columns = ['价格', '销量']
bid.sort_values('价格', ascending=False, inplace=True)

plt.close()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(1,bn+1),bid.价格, label='全市总销量: {}'.format(quan))
for i in range(bn):
    ax.scatter(i+1, bid.ix[i,0], s=bid.ix[i,1]**.7*30, c='r', alpha=.2)
    
plt.legend(loc='upper right', prop=zh_font)
plt.ylim(0,500)
plt.xlim(0,bn+1)
plt.yticks(range(0,550,50))
plt.xticks(range(1,bn+1,1), bid.index, fontproperties=zh_font)
plt.tight_layout()
plt.show()

KTV.index = range(1,len(KTV)+1)
KTV
```
